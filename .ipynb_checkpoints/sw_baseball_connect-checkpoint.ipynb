```python
import pandas as pd
import psycopg2 
import sqlalchemy as sa
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Use sqlalchemy to create an `engine`; the `engine` will be used by pandas to connect to postgreSQL
- the engine needs a connection defined in this format: `postgresql://user:password@server:port/database_name`
- everything after the `//` is specific to your database

```python
engine = sa.create_engine('postgresql://postgres:postgres@localhost:5432/lahman_baseball')
```

#### Now you can use the pandas `read_sql( )` method
- set `sql=` the query you want to run
- set `con = engine` to execute a query on any table in the connected database  

```python
players = pd.read_sql(sql = 'SELECT * FROM people;', con = engine)
players.head()
```

```python
players.columns.values
```

### Analyze all the colleges in the state of Tennessee. Which college has had the most success in the major leagues. Use whatever metric for success you like - number of players, number of games, salaries, world series wins, etc.

```python
schools = pd.read_sql(sql = 'SELECT * FROM schools;', con = engine)
schools.head(2)
```

```python
college_play = pd.read_sql(sql = 'SELECT * FROM collegeplaying;', con = engine)
college_play.head(2)
```

```python
tn_schools = schools.loc[schools.schoolstate == 'TN']
tn_schools.shape
```

```python
tn = pd.merge(tn_schools, college_play, on = 'schoolid')
tn.shape
```

```python
tn.head()
```

```python
tn_players = pd.merge(tn, players, on = 'playerid')
print(tn_players.shape)
tn_players.head(2)
```

#### For starters, let's look at total salary and find the top 5

```python
salaries = pd.read_sql(sql = 'SELECT * FROM salaries;', con = engine)
salaries.head(2)
```

```python
tn_salaries = pd.merge(tn_players, salaries, on = 'playerid')
```

```python
tn_salaries.groupby(['playerid', 'namefirst', 'namelast'])['salary'] \
          .agg('sum') \
          .to_frame() \
          .sort_values('salary', ascending = False) \
          .head()
```

#### What about All Star game appearances?

```python
allstar = pd.read_sql(sql = 'SELECT * FROM allstarfull;', con = engine)
allstar.head(2)
```

```python
tn_allstars = pd.merge(tn_players, allstar, on = 'playerid')
```

```python
tn_allstars.groupby(['playerid', 'namefirst', 'namelast'])['schoolid']\
          .agg('count') \
          .reset_index() \
          .sort_values('schoolid', ascending = False) \
.rename(columns = {'schoolid': 'count'})
          
```

### Is there any correlation between number of wins and team salary? Use data from 2000 and later to answer this question. As you do this analysis, keep in mind that salaries across the whole league tend to increase together, so you may want to look on a year-by-year basis.

- first we'll get salaries

```python
salaries2000 = salaries.loc[salaries.yearid >= 2000]
salaries2000.head(2)
```

```python
team_salaries = salaries2000.groupby(['yearid','teamid']) \
                            .sum() \
                            .reset_index() 
team_salaries.head(2)
```

- next we'll get number of wins

```python
teams = pd.read_sql(sql = 'SELECT * FROM teams;', con = engine)
teams.head(2)
```

```python
team_wins = teams[['yearid', 'teamid', 'w']]
team_wins = team_wins.loc[team_wins.yearid >= 2000]
team_wins.head(2)
```

```python
salaries_and_wins = pd.merge(team_salaries, team_wins, on = ['yearid', 'teamid'])
salaries_and_wins.head(2)
```

```python
salaries_and_wins.groupby('yearid').corr()
```

```python
g = sns.FacetGrid(salaries_and_wins, col = 'yearid', col_wrap = 5, height = 2)
g.map(plt.scatter, 'salary', 'w')
plt.show()
```

### In this question, you will explore the connection between number of wins and attendance.

 - Does there appear to be any correlation between attendance at home games and number of wins?
 - Do teams that win the world series see a boost in attendance the following year? What about teams that made the playoffs? Making the playoffs means either being a division winner or a wild card winner.

```python
team_wins_all = teams[['yearid', 'teamid', 'w']]
team_wins_all.head()
```

```python
homegames = pd.read_sql(sql = 'SELECT * FROM homegames;', con = engine)
homegames.head(2)
```

```python
homegames = homegames[['year', 'team', 'games', 'attendance']]
homegames['avg_attendance'] = homegames.attendance / homegames.games
homegames.head(2)
```

```python
wins_and_attendance = pd.merge(team_wins_all, homegames,
                               left_on = ['yearid', 'teamid'],
                               right_on = ['year', 'team'])
wins_and_attendance.head()
```

```python
plt.scatter('w', 'attendance', data = wins_and_attendance)
plt.xlabel('wins')
plt.ylabel('attendance')
plt.show()
```
